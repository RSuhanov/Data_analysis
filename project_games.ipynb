Расчитаем соотношение NaN-значений с общим количеством записей в столбцах, посчитаем средний % NaN-значений в датасете.

```python
def percent_isnull(row): 
    value =  (row.isnull().sum() / len(row) * 100).sort_values(ascending = False).round(1)
    value_mean = value.mean()
    display(value)
    print('Cредний процент:', int(value_mean), '%')
    return

percent_isnull(data)
```

В столбцах Critic_Score, Rating, User_Score от 40 до 51% NaN-значений, что может негативно повлиять на корректность расчетов и исказить результаты анализа. 

Проверим за какой период отсутствуют данные.

```python
data_nan = data.query('Critic_Score.isna() or Rating.isna() or User_Score.isna()')
data_nan_year = data_nan.pivot_table(index = 'Year_of_Release', values = 'Platform', aggfunc = 'count')
display(data_nan_year.head())
display(data_nan_year.tail())
```

NaN-значения столбцах Critic_Score, Rating, User_Score встречаются с 1980 по 2016. Можно предположить, что данные были утеряны/искажены в результате сбоя ПО/СУБД или данные были агрегированы из разных источников, хранивших информацию о рейтингах в разных форматах, либо хранивших неполную информацию, поэтому при слиянии появились NaN-значения. 

Удалить из датасета 40-50% данных от общего объема или занулить/усреднить рейтинги в данном исследовании некорректно, т.к. может исказить результаты корреляционного анализа и стат-тестов, поэтому предлагаемый метод решения - замена NaN-значений на "цифровую заглушку" (-1), чтобы в дальнейшем можно были использовать фильтрацию в запросах.

```python
data_tbd = data.query('User_Score == "tbd"')
data_tbd_year = data_tbd.pivot_table(index = 'Year_of_Release', values = 'Platform', aggfunc = 'count')
display(data_tbd_year.head())
display(data_tbd_year.tail())
```

В столбце User_Score помимо NaN-значений содержится порядка 15%  значений - "tbd". Данная аббревиатура расшифровывается как "to be determined", т.е. игра не набрала достаточное количество оценок. 

Удалить из датасета порядка 15% данных от общего объема или занулить/усреднить пользовательский рейтинг, при отсутствии достаточного количества оценок, в данном исследовании будет некорректно, т.к. может исказить результаты анализа, поэтому предлагаемый метод решения - замена tbd на "цифровую заглушку" (-1), чтобы в дальнейшем можно были использовать фильтрацию в запросах.

Можно заключить, что:  
- до 1995 года количество ежегодно выпускаемых игр находится примерно на одном уровне;

Резкое, кратное увеличение выпускаемых игр примерно каждые 5 лет связано с выпуском новых моделей процессоров и видеокарт, производительность которых возрастает примерно в 2 - 4 раза, что позволяет разработчикам игр значительно улучшать качество продукта, тем самым стимулируя спрос и предложение.

### Определение актуального периода 

Для определения актуального периода исследуемых данных необходимо понять, за какой характерный срок появляются новые и исчезают старые платформы. 


```python
data_platform_life_time = data.pivot_table(index = 'platform', values = 'year_of_release', aggfunc = ['min', 'max'])
data_platform_life_time['platform_life_time'] = data_platform_life_time[('max', 'year_of_release')] \
                                              - data_platform_life_time[('min', 'year_of_release')]
top_5_platform_life_time = data_platform_life_time.query('platform in @top_5_platform.index')
display(top_5_platform_life_time.sort_values(by = 'platform_life_time', ascending = False).head())
platform_years = top_5_platform_life_time['platform_life_time'].median()
print(f'Среднее время жизни ТОП-5 платформ: {platform_years:.1f} лет')
```

С учетом того, что среднее время жизни платформы составляет 11 лет, а рекламная компания планируется на 2017 год, то для получения результатов, релевантных целям исследования необходимо оставить данные за период с 2005 по 2016 год. Это и будет актуальным периодом.

### Влияние на продажи внутри популярных платформ отзывов пользователей и критиков

Определим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. 

Для этого построим диаграммы рассеяния зависимости между продажами ТОП-3 платформ и отзывами критиков, зависимости между продажами ТОП-3 платформ и оценками пользователей. 

Чтобы не исказить результаты анализа на этапе обработки данных была проведена замена NaN-значений на "цифровую заглушку" (-1). Поэтому сделаем выборку данных только с отзывами, оценками раными нулю и выше.  

```python
data_platform_score = data_actual.query('critic_score >=0 and user_score >=0')
ax = data_platform_score.query('platform == "PS4"').plot(x = 'critic_score', y = 'total_sales', figsize=(14,7), \
        title = 'Диаграмма рассеяния зависимости между продажами (млн. $) ТОП-3 платформ и отзывами критиков за период с 2005 до 2016', \
                                                    alpha=0.3, style = 'o', label ='PS4')
data_platform_score.query('platform == "XOne"').plot(x = 'critic_score', y = 'total_sales', style = 'o', \
                                               alpha=0.3, ax = ax, label ='XOne')
data_platform_score.query('platform == "3DS"').plot(kind = 'scatter', x = 'critic_score', y = 'total_sales', \
                                              alpha=0.3, color = 'green', ax = ax, label ='3DS')
plt.legend();
```

Можно заключить, что:

- с ростом рейтинга критиков растет и уровень продаж игр;
- зависимость наиболее выражена, если рейтинг превышает уровень 60;
- более качественные продукты получают более высокий рейтинг критиков, что оказывает влияние на потенциальных покупателей и продажи.  

```python
ax = data_platform_score.query('platform == "PS4"').plot(x = 'user_score', y = 'total_sales', \
    title = 'Диаграмма рассеяния зависимости между продажами (млн. $) ТОП-3 платформ и оценками пользователей за период с 2005 до 2016',\
                                                    alpha=0.3, figsize=(14,7), style = 'o', label ='PS4')
data_platform_score.query('platform == "XOne"').plot(x = 'user_score', y = 'total_sales', style = 'o', \
                                                alpha=0.3, ax = ax, label ='XOne')
data_platform_score.query('platform == "3DS"').plot(kind = 'scatter', x = 'user_score', y = 'total_sales', \
                                               color = 'green', alpha=0.3, ax = ax, label ='3DS')
plt.legend();
```

## Проверка гипотез

Сформулируем гипотезы, для проверки которых применим  метод ttest_ind. Определим пороговое значение alpha = 0.05

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

 - Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
 - Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются.

```python
data_platform_xbox = data_actual.query('platform == "XOne" and user_score >= 0')
data_platform_pc = data_actual.query('platform == "PC" and user_score >= 0')
```

```python
alpha = 0.05
results = st.ttest_ind(data_platform_xbox['user_score'], data_platform_pc['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вывод

- значение p-value говорит о том, что полученный результат является статистически значимым и разница не случайна, поэтому нулевая гипотеза отвергается.

<font size="2">([к содержанию](#Содержание))</font>

### Средние пользовательские рейтинги жанров Action и Sports разные

 - Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые;
 - Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются.

```python
data_genre_action = data_actual.query('genre == "Action" and user_score >= 0')
data_genre_sport = data_actual.query('genre == "Sports" and user_score >= 0')
```

```python
alpha = 0.05
results = st.ttest_ind(data_genre_action['user_score'], data_genre_sport['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вывод

- значение p-value говорит о том, что полученный результат является статистически значимым и разница не случайна, поэтому нулевая гипотеза отвергается.
