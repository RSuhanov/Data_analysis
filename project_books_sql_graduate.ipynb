## Задания

Приступим к выполнению поставленных задач.

### Расчет количества книг,  вышедших после 1 января 2000 года

```python
query='''
SELECT 
     COUNT(book_id)
FROM books
WHERE publication_date > '2000-01-01'
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

**Решение:**  после 1 января 2000 года вышло 819 книг.

### Расчет количества обзоров и средней оценки для каждой книги

```python
query='''
WITH 
avg_rating AS
  (SELECT b.book_id,
           title,
           AVG(r.rating) AS avg_rating
    FROM books b
    LEFT JOIN ratings r ON r.book_id = b.book_id
    GROUP BY b.book_id),
    
cnt_reviews AS
   (SELECT b.book_id,
          title,
          COUNT(rev.review_id) AS cnt_reviews
    FROM books AS b
    LEFT JOIN reviews rev ON rev.book_id = b.book_id
    GROUP BY b.book_id)

SELECT a.book_id,
       a.title, 
       c.cnt_reviews,
       a.avg_rating
FROM avg_rating AS a
JOIN cnt_reviews c ON a.book_id = c.book_id
ORDER BY a.avg_rating DESC
                                   
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

**Решение:** 
выполнен расчет количества обзоров и средней оценки для каждой книги, результат в итоговой таблице отсортирован по убыванию рейтинга.

### Определение издательства,  выпустившего наибольшее количество книг толще 50 страниц

```python
query='''
SELECT 
       p.publisher_id,
       p.publisher,
       COUNT(book_id) AS cnt_books
FROM books AS b
JOIN publishers p ON p.publisher_id = b.publisher_id
WHERE num_pages > 50
GROUP BY p.publisher_id,
         p.publisher
ORDER BY cnt_books DESC
LIMIT 1
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

**Решение:** определено издательство - Penguin Books, выпустившее наибольшее количество книг толще 50 страниц (из анализа исключены брошюры). 

### Определение автора с самой высокой средней оценкой книг

```python
query='''
SELECT 
       a.author_id,
       a.author,
       b.book_id,
       b.title,
       AVG(r.rating) AS avg_rating
FROM books AS b
LEFT JOIN authors a ON a.author_id = b.author_id
LEFT JOIN ratings r ON r.book_id = b.book_id
GROUP BY 
         b.book_id,
         b.title,
         a.author_id,
         a.author
HAVING
     COUNT(r.rating_id) >= 50
ORDER BY avg_rating DESC
LIMIT 1;
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

### Расчет среднего количества обзоров от пользователей, которые поставили больше 50 оценок

```python
query='''
WITH mean_users_review 
AS (SELECT r.username
   FROM ratings r
   GROUP BY r.username
   HAVING (COUNT(r.rating_id) > 50))
SELECT AVG(t.cnt_review) AS avg_review
FROM
  (SELECT COUNT(rev.review_id) AS cnt_review
   FROM reviews AS rev
   WHERE rev.username in
        (SELECT *
        FROM mean_users_review)
   GROUP BY rev.username) AS t
'''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

**Решение:** пользователи, поставившие больше 50 оценок в среднем сделали 24 обзора.
