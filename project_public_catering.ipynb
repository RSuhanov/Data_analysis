### Подготовка к анализу

Выполним дополнительную подготовку данных к анализу, в первую очередь, поработаем с неявными дубликатами в наименованиях. 

```python
for replace_value in ['семейное', 'семейный', 'бар', 'кальян- бар', 'закусочная', \
                      'кальянная', 'пиццерия', 'пицерия', 'ресторан', 'предприятие быстрого питания', \
                      'предприятие быстрого обслуживания', 'корчма', 'магазин готовой еды', 'готовые блюда', \
                      '"', '«', '»', '-']:

    data['object_name'] = data['object_name'].str.replace(replace_value, '')
    data['object_name'] = data['object_name'].str.lstrip()
```

```python
search_duplicates = data.query('chain == "да"').pivot_table(index='object_name',                             
                                    values='id',
                                    aggfunc='count').reset_index()
search_duplicates = search_duplicates.sort_values(by = 'id', ascending = False)
```

Определим наиболее крупные сети объектов питания.  

```python
display(search_duplicates.head(15))
```

Проведем визуальный анализ уникальных наименований для минимизации неявных дубликатов. 

```python
display(search_duplicates['object_name'].sort_values().unique().tolist())
```

Присутствует большое количество неявных дубликатов в наименованиях объектов. Вероятно в ПО отсутствует необходимый функционал контроля при добавлении новых записей / использования предзаполненных справочников либо данные могли быть агрегигорованы из нескольких источников без дополнительной обработки. 

С помощью функции поработаем с дубликатами в наименованиях объектов наиболее крупных сетей. 

```python
def search_duplicates(duplicate, replacement):
    duplicate_meaning = data[data['object_name'].str.contains(duplicate)]['object_name'].unique().tolist()
    data.loc[data['object_name'].isin(duplicate_meaning), 'object_name'] = replacement
```

```python
search_duplicates('макд|macd|mcd', 'макдоналдс')  
search_duplicates('старбакс|starbucks', 'старбакс')
search_duplicates('kfc|кфс', 'кфс')
search_duplicates('burger king|бургер кинг', 'бургер кинг')
search_duplicates('шоколадница', 'шоколадница')
search_duplicates('теремок', 'теремок')
search_duplicates('крошка картошка', 'крошка картошка')
search_duplicates('милти', 'милти')
search_duplicates('суши wok', 'суши wok')
search_duplicates('чайхона №1', 'чайхона №1')
search_duplicates('тануки', 'тануки')
search_duplicates('якитория', 'якитория')
search_duplicates('якитория', 'якитория')
```

```python
search_duplicates = data.query('chain == "да"').pivot_table(index='object_name',                             
                                    values='id',
                                    aggfunc='count').reset_index()
search_duplicates = search_duplicates.sort_values(by = 'id', ascending = False)
```

Проверим полученный результат. 

```python
display(search_duplicates.head(15))
```

По возможности, данные нормализовали, в результате количество объектов у наиболее крупных сетей существенно увеличилось.

## Анализ данных

### Соотношение видов объектов общественного питания по количеству

Проведем исследование соотношения видов бъектов общественного питания по количеству. С помощью группировки данных в сводной таблице получим количество объектов общественного питания для каждого вида и построим по этим данным график.

```python
object_type_proportion = data.pivot_table(index='object_type',
                                        values='id',
                                        aggfunc='count').reset_index()
object_type_proportion['%'] = (object_type_proportion['id'] / object_type_proportion['id'].sum() * 100).round(1)
plot_object_type_proportion = object_type_proportion.sort_values(by = 'id', ascending = False)
display(plot_object_type_proportion)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
plt.bar(plot_object_type_proportion['object_type'], plot_object_type_proportion['id'],
        color= ['royalblue', 'teal', 'plum', 'cyan', 'palegreen', 'green', 'orange', 'gold', 'red'])
plt.title('Соотношение видов объектов общественного питания по количеству в Москве', size=17)
plt.xticks(rotation=45)
plt.xlabel('Вид заведения', size=13)
plt.ylabel('Количество заведений', size=13)
plt.show()
```

### Cоотношение сетевых и несетевых объектов по количеству

Проведем исследование соотношения сетевых и несетевых объектов. С помощью сводной таблицы получим количество объектов общественного питания  и построим по этим данным график.

```python
chain_type_proportion = data.pivot_table(index='chain',
                                        values='id',
                                        aggfunc='count').reset_index()
chain_type_proportion['%'] = (chain_type_proportion['id'] / chain_type_proportion['id'].sum() * 100).round(1)
plot_chain_type_proportion = chain_type_proportion.sort_values(by = 'id', ascending = False)
display(chain_type_proportion)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8, 6))
plt.bar(plot_chain_type_proportion['chain'], plot_chain_type_proportion['id'],
        color= ['royalblue', 'palegreen'])
plt.title('Соотношение сетевых и несетевых объектов общественного питания по количеству в Москве', size=17)
plt.xlabel('Сетевое заведение', size=13)
plt.ylabel('Количество заведений', size=13)
plt.show()
```

### Виды объектов общественного питания с характерным сетевым распространением

### Характерное для сетевых заведений

Исследуем, что наиболее характерно для сетевых заведений: 

- много заведений с небольшим числом посадочных мест в каждом;
- мало заведений с большим количеством посадочных мест.  



```python
object_chain_type = object_chain_type.groupby('object_name').agg({'id': 'count', 'number': 'mean'}).reset_index()
```

```python
display(object_chain_type['id'].describe())
```

```python
display(object_chain_type['number'].describe())
```

Категорируем данные по следующим группам:

- мало заведений (до 4);
- много заведений (более 4);
- мало посадочных мест (до 50);
- много посадочных мест (более 50).

и построим по этим данным диаграмму рассеяния. 

```python
object_chain_type['object_num_group'] = 'Мало заведений'
object_chain_type.loc[object_chain_type['id'] > 5, 'object_num_group'] = 'Много заведений'
object_chain_type['seat_num_group'] = 'Мало посадочных мест'
object_chain_type.loc[object_chain_type['number'] > 50, 'seat_num_group'] = 'Много посадочных мест'

object_chain_type['Группа'] = object_chain_type['object_num_group'].astype(str) + ' - ' + object_chain_type['seat_num_group'].astype(str)
plt.figure(figsize=(17,10))
sns.scatterplot(data = object_chain_type, x = 'id', y = 'number', hue = 'Группа')
plt.title('Распределение сетевых заведений общественного питания Москвы по группам в зависимости от количества посадочных мест и количества объектов', size=17)
plt.xlabel('Количество заведений', size=13)
plt.ylabel('Медианное количество посадочных мест', size=13)
plt.show()
```

```python
object_chain_type = object_chain_type.groupby('Группа').agg({'object_name': 'nunique', 'id': 'median', 'number': 'median'})\
                                   .rename(columns = {'object_name': 'Количество сетей', 'id': 'Медианное количество заведений', 'number': 'Медианное количество посадочных мест'})
object_chain_type = object_chain_type.sort_values(by = 'Количество сетей', ascending = False)

object_chain_type['%'] = (object_chain_type['Количество сетей'] / object_chain_type['Количество сетей'].sum() * 100).round(1)
display(object_chain_type)
```

**Вывод:**

Для сетевых заведений наиболее характерно:
- мало заведений / много посадочных мест - 48 %;
- мало заведений / мало посадочных мест - 32.5 %.

```python
data_filtered.groupby('object_type')['number'].describe().sort_values(by='50%', ascending = False)
```

```python
number_chain_type = data_filtered.query('chain == "да"')
number_chain_avg = number_chain_type.pivot_table(index='object_type',
                                      values='number',
                                      aggfunc='median').reset_index()
number_chain_avg = number_chain_avg.sort_values(by='number', ascending=False).reset_index(drop=True)
```

```python
number_non_chain_type = data_filtered.query('chain == "нет"')
number_non_chain_avg = number_non_chain_type.pivot_table(index='object_type',
                                      values='number',
                                      aggfunc='median').reset_index()
number_non_chain_avg = number_non_chain_avg.sort_values(by='number', ascending=False).reset_index(drop=True)
```

```python
chain_non_chain_avg = number_chain_avg.merge(
    number_non_chain_avg, left_on='object_type', right_on='object_type', \
    how='left', suffixes=['_chain', '_non_chain'])
display(chain_non_chain_avg)
```
