```python
def shaper(data, column):
    q1 = data[column].quantile(0.25)                 
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    shape = (data[column] >= (q1 - 1.5 * iqr)) & (data[column] <= (q3 + 1.5 * iqr))
    return data.loc[shape]
```

```python
shaper_colums_list = ['total_area', 'last_price', 'rooms', 'ceiling_height', 'days_exposition']

for i in range(len(shaper_colums_list)):
    data[shaper_colums_list[i]] = shaper(data, shaper_colums_list[i])[shaper_colums_list[i]]
```

```python
data['citycenters_nearest_km'] = round(data['citycenters_nearest'] / 1000)

def center_distance_grouped(row):
    center_distance = row['citycenters_nearest_km']
    if center_distance <= 1:
        return ('1 км')
    if 1 < center_distance <= 2:
        return ('2 км')
    if 2 < center_distance <= 3:
        return ('3 км')
    if 3 < center_distance <= 4:
        return ('4 км')
    if 4 < center_distance <= 5:
        return ('5 км')
    if 5 < center_distance <= 6:
        return ('6 км')
    if 6 < center_distance <= 7:
        return ('7 км')
    if 7 < center_distance <= 8:
        return ('8 км')
    if 8 < center_distance <= 9:
        return ('9 км')
    if 9 < center_distance <= 10:
        return ('10 км')
    if 10 < center_distance <= 11:
        return ('11 км')
    if 11 < center_distance <= 12:
        return ('12 км')
    if 12 < center_distance <= 13:
        return ('13 км')
    if 13 < center_distance <= 14:
        return ('14 км')
    if 14 < center_distance <= 15:
        return ('15 км')
    if 15 < center_distance <= 16:
        return ('16 км')
    if 16 < center_distance <= 17:
        return ('17 км')
    if 17 < center_distance <= 18:
        return ('18 км')
    if 18 < center_distance <= 19:
        return ('19 км')
    if 19 < center_distance <= 20:
        return ('20 км')
    if 20 < center_distance <= 21:
        return ('21 км')
    if 21 < center_distance <= 22:
        return ('22 км')
    if 22 < center_distance <= 23:
        return ('23 км')
    if 23 < center_distance <= 24:
        return ('24 км')
    if 24 < center_distance <= 25:
        return ('25 км')
    if center_distance > 25:
        return ('от 25 км и более')
    else:
        return 'неизвестно'
    
data['citycenters_nearest_grouped'] = data.apply(center_distance_grouped, axis = 1)
```
