**Описание проекта**

Вы аналитик компании «Мегалайн» — федерального оператора сотовой связи. Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Вам предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В вашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год.

<font size="4"><b>Описание тарифов</b></font><br/><br/>

**Тариф «Смарт»**
 - Ежемесячная плата: 550 рублей
 - Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
 - Стоимость услуг сверх тарифного пакета: минута разговора - 3 рубля («Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута); сообщение - 3 рубля; 1 Гб интернет трафика - 200 рублей.

**Тариф «Ультра»**
 - Ежемесячная плата: 1950 рублей
 - Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
 - Стоимость услуг сверх тарифного пакета: минута разговора - 1 рубль; сообщение - 1 рубль; 1 Гб интернет трафика - 150 рублей.
 

**Примечание:**
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.

Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

По условиям своих тарифов "Мегалайн» округляет мегабайты до гигабайт, а в случае превышения объема интернет-трафика оплата производится за каждый гигабайт. Для расчета создадим дополнительный столбец, выполним необходимую обработку данных.

```python
def percent_isnull(row): 
    value =  (row.isnull().sum() / len(row) * 100).sort_values(ascending = False).round(1)
    value_mean = value.mean()
    display(value)
    print('Cредний процент:', int(value_mean), '%')
    return

percent_isnull(data)
```

Выполним расчет выручки с помощью функции, результаты сохраним в дополнительном столбце revenue. 

```python
def month_revenue(row):
    monthly_fee = row['rub_monthly_fee']
    if (row['internet_total_in_month'] - row['gb_per_month_included']) > 0:
        internet_fee = (row['internet_total_in_month'] - row['gb_per_month_included']) * (row['rub_per_gb'])
    else: 
        internet_fee = 0
    if (row['messages_total_in_month'] - row['messages_included']) > 0:
        messages_fee = (row['messages_total_in_month'] - row['messages_included']) * (row['rub_per_message'])
    else:
        messages_fee = 0
    if (row['month_calls_duration'] - row['minutes_included']) > 0:
        calls_fee = (row['month_calls_duration'] - row['minutes_included']) * (row['rub_per_minute'])
    else:
        calls_fee = 0
    month_revenue_user = monthly_fee + internet_fee + messages_fee + calls_fee
    return month_revenue_user    
```

```python
data['revenue'] = data.apply(month_revenue, axis =1)
```

## Проверка гипотез

Далее приступим с формированию и проверке гипотез.

### Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различается


Поскольку будем исследовать методом  ttest_ind, то необходимо убрать выбросы для корректного сравнения средних значений независимых выборок.  

```python
def shaper(data, column):
    q1 = data[column].quantile(0.25)                 
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    shape = (data[column] >= (q1 - 1.5 * iqr)) & (data[column] <= (q3 + 1.5 * iqr))
    return data.loc[shape]
```

Выполним фильтрацию данных, проверим полученные результаты и сделаем выводы. 

```python
data_smart['revenue'].describe()
```

```python
data_smart = shaper(data_smart, 'revenue')
```

```python
data_smart['revenue'].describe()
```

```python
data_ultra['revenue'].describe()
```

```python
data_ultra = shaper(data_ultra, 'revenue')
```

```python
data_ultra['revenue'].describe()
```

Убрали выбросы, приступаем к анализу. 

Определим значение критического уровня статистической значимости (alpha) = 0.05.

 - H0 (нулевая гипотеза): выручка пользователей тарифов «Ультра» и «Смарт» не различается
 - H1 (альтернативная гипотеза): выручка пользователей тарифов «Ультра» и «Смарт» различается


```python
alpha = .05 
results = st.ttest_ind(data_ultra['revenue'], data_smart['revenue'], equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вывод
- значение p-value говорит о том, что полученный результат является статистически значимым и разница не случайна, поэтому нулевая гипотеза отвергается. 

###  Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Для проверки гипотезы необходимо провести дополнительную обработку данных и сформировать из общей выборки два сегмента: 
- пользователи из Москвы;
- пользователи из регионов. 



```python
data['region'] = np.where(data['city'] == 'Москва', 'Москва', 'Регионы')
```

```python
data_moscow = data.query('region == "Москва"')
display(data_moscow.sample(15))
```

```python
data_regions = data.query('region != "Москва"')
display(data_regions.sample(15))
```

```python
data_moscow['revenue'].describe()
```

```python
data_moscow = shaper(data_moscow, 'revenue')
```

```python
data_moscow['revenue'].describe()
```

```python
data_regions['revenue'].describe()
```

```python
data_regions = shaper(data_regions, 'revenue')
```

```python
data_regions['revenue'].describe()
```

Определим значение критического уровня статистической значимости (alpha) = 0.05.

- H0 (нулевая гипотеза): средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов
- H1 (альтернативная гипотеза): средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

```python
alpha = .05 
results = st.ttest_ind(data_moscow['revenue'], data_regions['revenue'], equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

С вероятностью 30% такой результат может быть получен случайно. 

Перепроверим полученный результат.

```python
print('Cредняя выручка пользователей из Москвы в месяц: {:.1f}'.format(data_moscow['revenue'].mean()))
print('Средняя выручка пользователей из других регионов в месяц: {:.1f}'.format(data_regions['revenue'].mean()))
```

В Москве клиенты в среднем платят на 35 руб. больше, чем в регионах, что не является статистически значимой величной в данном исследовании. 

Сформулированные в рамках исследования гипотезы проверены, рузультаты получены.
