Требуется помочь найти самых неэффективных операторов. 

Признаки низкой эффективности:
- много пропущенных входящих вызовов;
- долгое ожидание ответа при входящих внешних вызовах;
- мало исходящих вывозов — в случае тех колл-центров, что специализируются на обзвонах, а не обработке входящих звонков.

**Необходимо:**
- провести исследовательский анализ данных;
- определить неэффективных операторов;
- проверьте статистические гипотезы;
- подготовить презентацию;
- подготовить дашборд.

Варианты наборов для дашборда:
  - набор №1
     - гистограмма распределения длительности звонков;
     - круговая диаграмма, отображающая соотношение количества внутренних и внешних вызовов;
     - фильтр по направлению вызовов.
  - набор №2
      - диаграмма количества вызовов по дням;
      - круговая диаграмма, отображающая соотношение количества входящих и исходящих вызовов;
      - фильтр дашборда по типу вызова (внешний или внутренний).

## Описание данных

Предоставлены датасеты, описывающие использование услуг провайдера виртуальный телефонии. Его клиенты — колл-центры, которые: 
- распределяют входящие вызовы на операторов; 
- совершают исходящие вызовы силами операторов. 

Примечание: операторы могут делать внутренние вызовы т.е. вызовы друг между другом внутри сети виртуальной телефонии.

**Колонки в telecom_dataset.csv:**
- `user_id` — Идентификатор клиентского аккаунта в сервисе

- `date` — Дата статистики

- `direction` — Направление вызовов (out - исходящий вызов, in — входящий вызов)

- `internal` — Является ли звонок внутренним звонком между операторами клиента

- `operator_id` — Идентификатор оператора

- `is_missed_call` — Является ли звонок пропущенным

- `calls_count` — Количество звонков

- `call_duration` — Длительность звонка (без учета времени ожидания)

- `total_call_duration` — Длительность звонка (с учетом времени ожидания)

**Колонки в telecom_clients.csv:**
- `user_id` — Идентификатор клиентского аккаунта в сервисе

- `tariff_plan` — Текущий тарифный план клиента

- `date_start` — Дата регистрации клиентв в сервисе

```python
def percent_null_value(row): 
    value =  (row.isna().sum() / row.shape[0] * 100).round(2)
    print('Доля строк с нулевым значением:', value, '%')
    return
```

Критерием категоризации клиентов будем считать разницу более чем 30% между долями входящих / исходящих вызовов.  

```python
def user_categorization (data):
    if (data['in'] !=0 or data['out'] !=0): 
        if  (data['% in total'] / data['% out total']) > 1.3:
            data['category'] = 'принимающий'
        elif (data['% in total'] / data['% out total']) < 0.7:
             data['category'] = 'обзвон'
        else:
             data['category'] = 'многофункциональный'
    elif (data['in'] !=0 & data['out'] !=0): 
          data['category'] = 'не определена'
    return data 
```

## Определение неэффективных операторов

Для категорирования операторов применим функцию на основе сформированных границ неэффективности.   

```python
dataset_missed_call_grouped.head()
```

```python
def operator_categorization (data, column, kpi, direction):
    if direction == 'in':
        if data[column] > kpi:
            return 'неэффективный'
        elif data[column] < kpi:
            return 'эффективный'
      
    if direction == 'out':
        if data[column] < kpi:
            return 'неэффективный'
        elif data[column] > kpi:
            return 'эффективный'
    else:
            return 'не определено'
```

## Проверка гипотез

Для подтверждения гипотез о равенстве средних двух генеральных совокупностей применим Т-тест. Уровень статистической значимости примем равным - 5%.

```python
alpha = 0.05
```

### Среднее количество пропущенных входящих вызовов зависит от тарифного плана

Сформулируем гипотезы:

- нулевая (H₀) - среднее количество пропущенных входящих вызовов не зависит от тарифного плана;
- альтернативная (H₁) - среднее количество пропущенных входящих вызовов  зависит от тарифного плана.

```python
dataset_missed_call_tariff_a = dataset_missed_call.query('tariff_plan == "A"')['calls_count']
dataset_missed_call_tariff_a.count()
```

```python
dataset_missed_call_tariff_b = dataset_missed_call.query('tariff_plan == "B"')['calls_count']
dataset_missed_call_tariff_b.count()
```

```python
dataset_missed_call_tariff_c = dataset_missed_call.query('tariff_plan == "C"')['calls_count']
dataset_missed_call_tariff_c.count()
```

```python
results = st.ttest_ind(dataset_missed_call_tariff_a, dataset_missed_call_tariff_b)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
results = st.ttest_ind(dataset_missed_call_tariff_a, dataset_missed_call_tariff_c)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
results = st.ttest_ind(dataset_missed_call_tariff_b, dataset_missed_call_tariff_c)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Согласно полученным результатам, при уровене статистической значимости - 5% между тарифными планами отсутствуют значимые различия средних значений количества пропущенных входящих вызовов.
