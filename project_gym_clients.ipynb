#  Прогноз вероятности оттока  клиентов сети  фитнес-центров и  подготовка рекомендаций по их удержанию.

## Описание проекта

Сеть фитнес-центров разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных. Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Отдел по работе с клиентами перевёл в электронный вид множество клиентских анкет. 

**Необходимо провести анализ и подготовить рекомендации по удержанию клиентов:**
- спрогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
   - 1) выделить целевые группы клиентов;
   - 2) предложить меры по снижению оттока;
   - 3) определить другие особенности взаимодействия с клиентами.

## Описание данных

Заказчик подготовил данные на месяц до оттока и факт оттока на определённый месяц. Набор данных включает следующие поля:

**Данные клиента за предыдущий до проверки факта оттока месяц:**
- `'gender'` — пол ("1" - мужской, "0" - женский);

- `'Near_Location'` — проживание или работа в районе, где находится фитнес-центр;

- `'Partner'` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);

- `Promo_friends` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);

- `'Phone'` — наличие контактного телефона;

- `'Age'` — возраст;

- `'Lifetime'` — время с момента первого обращения в фитнес-центр (в месяцах).

**Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:**
- `'Contract_period'` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);

- `'Month_to_end_contract'` — срок до окончания текущего действующего абонемента (в месяцах);

- `'Group_visits'` — факт посещения групповых занятий;

- `'Avg_class_frequency_total'` — средняя частота посещений в неделю за все время с начала действия абонемента;

- `'Avg_class_frequency_current_month'` — средняя частота посещений в неделю за предыдущий месяц;

- `'Avg_additional_charges_total'` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

- `'Churn'`  — факт оттока в текущем месяце.

## Построение  модели прогнозирования оттока клиентов

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце. Разделим датасет в пропорции 80% на 20% на обучающую и валидационную выборку функцией train_test_split() и стандартизируем полученные данные.

```python
X = data.drop('churn', axis = 1)
y = data['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

Обучим модель на train-выборке двумя способами: "логистической регрессией" (LogisticRegression), "случайным лесом" (RandomForestClassifier). 

Качество полученных моделей оценим по метрикам:
- Accuracy (доля правильных ответов);
- Precision (точность);
- Recal (полнота).

###  Логистическая регрессия

```python
model = LogisticRegression()
```

```python
model.fit(X_train_st, y_train)
```

```python
probabilities = model.predict_proba(X_test_st)[:,1]
predictions = model.predict(X_test_st)
```

```python
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
```

### Cлучайный лес

```python
model = RandomForestClassifier(random_state=0)
```

```python
model.fit(X_train_st, y_train)
```

```python
probabilities = model.predict_proba(X_test_st)[:,1]
predictions = model.predict(X_test_st)
```

```python
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
```

### Оценка метрик

На основе полученных результатов можно заключить, что:
- по метрике Accuracy (доля правильных ответов) обе модели показали схожие результаты;
- по метрике Precision (точность): "случайный лес" чуть лучше; 
- по метрике Recall (полнота) "логическая регрессия" существенно лучше.

##  Кластеризация клиентов

Для получения корректных результатов приведем данные к виду стандартного нормального распределения. 

```python
scaler = StandardScaler()
X_st = scaler.fit_transform(X)
```

Выполним кластеризацию клиентов (разделение на группы при отсутствии заранее заданных классов или "правильных ответов").

```python
linked = linkage(X_st, method = 'ward')
```

```python
plt.figure(figsize=(15, 10))
plt.title('Кластеризация клиентов фитнес-центра')
dendrogram(linked, orientation='top')
plt.show()
```

Исходя из полученных результатов можно выделить 4 кластера клиентов фитнес-центра.  

Далаее обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. По договоренности число кластеров (n) примем равным 5. 

```python
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(X_st)
data['cluster'] = labels
```

Посмотрим с помощью метода groupby() на средние значения признаков в полученных кластерах. 

```python
data.groupby('cluster').mean().T.round(3)
```

Для каждого полученного кластера посчитаем точную долю оттока клиентов с помощью метода groupby(). 

```python
print('Объем оттока клиентов фитнес-центра:', data['churn'].sum())
print('Доля оттока клиентов фитнес-центра:', ((data['churn'].sum()/data['churn'].count() * 100).round(2)),'%')
```

```python
cluster_count = (data
                     .groupby('cluster')
                     .agg({'churn':'count'})
                     .rename(columns = {'churn': 'count'})
                     .reset_index()
                     .sort_values(by='cluster', ascending=True)
                     .reset_index(drop=True))
display(cluster_count)
```

```python
cluster_churn_count = (data
                        .groupby('cluster')
                        .agg({'churn':'sum'})
                        .reset_index()
                        .sort_values(by='cluster', ascending=True)
                        .reset_index(drop=True)
                   )
cluster_churn_count['% churn of cluster'] = (cluster_churn_count['churn'] / cluster_count['count'] * 100).round(2)
cluster_churn_count['% churn of total'] = (cluster_churn_count['churn'] / cluster_churn_count['churn'].sum() * 100).round(2)
cluster_churn_count['cluster count'] = cluster_count['count']
cluster_churn_count_rate = cluster_churn_count.sort_values(by='% churn of cluster', ascending = False)
display(cluster_churn_count_rate)
```
